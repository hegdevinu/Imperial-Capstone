```python
import zipfile
import os

# 1. Unzip the file
zip_filename = 'Initial_data_points_starter.zip'  # Replace with your actual filename
extract_path = 'capstone_data'

with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
    print(f"--- Files extracted to: {extract_path} ---\n")

# 2. Map out the folder structure
for root, dirs, files in os.walk(extract_path):
    level = root.replace(extract_path, '').count(os.sep)
    indent = ' ' * 4 * (level)
    print(f"{indent}{os.path.basename(root)}/")
    sub_indent = ' ' * 4 * (level + 1)
    for f in files:
        if f.endswith('.npy'):
            print(f"{sub_indent}📄 {f}")
```

```python
import numpy as np
import os

base_path = 'capstone_data/initial_data/'

# We will loop through 1 to 8 to check each folder
for i in range(1, 9):
    folder_name = f'function_{i}'
    inputs_path = os.path.join(base_path, folder_name, 'initial_inputs.npy')
    outputs_path = os.path.join(base_path, folder_name, 'initial_outputs.npy')
    
    # Check if files exist before loading
    if os.path.exists(inputs_path) and os.path.exists(outputs_path):
        inputs = np.load(inputs_path)
        outputs = np.load(outputs_path)
        
        print(f"--- {folder_name.upper()} ---")
        print(f"Input Shape:  {inputs.shape}  (Rows: {inputs.shape[0]}, Dimensions: {inputs.shape[1]})")
        print(f"Output Shape: {outputs.shape}")
        # Show the first 2 rows just to see the type of numbers
        print(f"First 2 input rows:\n{inputs[:2]}")
        print("-" * 30 + "\n")
    else:
        print(f"Warning: Data for {folder_name} not found at {inputs_path}")
```

```python
import os
import json

# Define the structure
base_dir = "Notebooks"
if not os.path.exists(base_dir):
    os.makedirs(base_dir)

# Basic Notebook Template (JSON format)
template = {
 "cells": [
  {"cell_type": "markdown", "metadata": {}, "source": ["# Optimization for Function X\n", "Goal: Maximise the black-box function using Bayesian Optimization."]},
  {"cell_type": "code", "execution_count": None, "metadata": {}, "outputs": [], "source": [
      "import numpy as np\n",
      "import os\n",
      "# Load Data\n",
      "fn_num = 'X'\n",
      "path = f'../capstone_data/initial_data/function_{fn_num}/'\n",
      "X = np.load(os.path.join(path, 'initial_inputs.npy'))\n",
      "y = np.load(os.path.join(path, 'initial_outputs.npy'))\n",
      "print(f'Function {fn_num} loaded. Input shape:', X.shape)"
  ]}
 ],
 "metadata": {"kernelspec": {"display_name": "Python 3", "language": "python", "name": "python3"}},
 "nbformat": 4, "nbformat_minor": 5
}

# Create 8 notebooks
for i in range(1, 9):
    filename = os.path.join(base_dir, f"{i:02d}_Function_{i}.ipynb")
    # Customise the template for each specific function
    current_temp = json.loads(json.dumps(template).replace('\"X\"', str(i)))
    with open(filename, 'w') as f:
        json.dump(current_temp, f)

print("Success! Check your file navigator for the 'Notebooks' folder.")
```
