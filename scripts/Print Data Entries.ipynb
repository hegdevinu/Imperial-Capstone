```python
import numpy as np
import os

base_path = 'capstone_data/initial_data/'

for i in range(1, 9):
    folder_name = f'function_{i}'
    inputs_path = os.path.join(base_path, folder_name, 'initial_inputs.npy')
    outputs_path = os.path.join(base_path, folder_name, 'initial_outputs.npy')
    
    if os.path.exists(inputs_path) and os.path.exists(outputs_path):
        inputs = np.load(inputs_path)
        outputs = np.load(outputs_path)
        
        print(f"\n{'='*30} FUNCTION {i} {'='*30}")
        print(f"{'Row':<5} | {'Input Coordinates (X)':<65} | {'Result (y)'}")
        print("-" * 110)
        
        for idx in range(len(inputs)):
            # Format the input array for readability
            input_str = np.array2string(inputs[idx], precision=6, separator=', ')
            print(f"{idx:<5} | {input_str:<65} | {outputs[idx]:.10f}")
    else:
        print(f"\n[!] Data for Function {i} missing at: {inputs_path}")
```
